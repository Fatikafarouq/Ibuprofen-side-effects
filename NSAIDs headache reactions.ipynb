```python
import pandas as pd
import numpy as np

drug = pd.read_csv("DRUG25Q2.txt", delimiter = "$", encoding = "latin-1", low_memory= False)
reac = pd.read_csv("REAC25Q2.txt", delimiter = "$", encoding = "latin-1", low_memory= False)

headache_reactions = reac[reac["pt"] == "Headache"]
nsaid_drugs = ["IBUPROFEN", "ASPIRIN", "NAPROXEN", "DICLOFENAC", "CELECOXIB"]

results_list = []
for drug_name in nsaid_drugs:
    drug_df = drug[drug["drugname"] == drug_name.upper()]
    merged_df = pd.merge(drug_df, headache_reactions, on='primaryid')
    report_count = merged_df.shape[0]
    results_list.append({'drug': drug_name, 'side effect': 'Headache', 'number of reports': report_count})

results_df = pd.DataFrame(results_list)

print(results_df)
```
